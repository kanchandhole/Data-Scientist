**Q4.** Why LookupError class is used? Explain with an example KeyError and IndexError.

**Ans:**LookupError is the base class for errors that occur when a key or index is not found in a collection.

```python
d = {"a": 1}
print(d["b"])   # KeyError
```

```python
lst = [10, 20, 30]
print(lst[5])   # IndexError
```

**Q6.** List down some best practices for exception handling in python.

**ans:**
1. Catch specific exceptions, not generic except:

2. Use finally for cleanup operations (closing files, connections).

3. Avoid using exceptions for normal program flow.

4. Add meaningful error messages for clarity.

5. Log exceptions instead of only printing them.

6. Create custom exceptions for application-specific issues.

7. Keep try blocks short and focused.

8. Use else block when code should run only if no exception occurs.
