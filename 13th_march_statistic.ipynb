**Q1.** Explain the assumptions required to use ANOVA and provide examples of violations that could impact
the validity of the results.

**Ans:**

Assumptions Required to Use ANOVA
1. Independence of Observations

Explanation: The data collected from each group must be independent of one another. One observation should not influence another.

Violation Example: Measuring the same studentsâ€™ test scores multiple times but treating them as separate observations. This can lead to misleading significance results.

2. Normality

Explanation: The dependent variable should be approximately normally distributed within each group.

Violation Example: If one groupâ€™s data is highly skewed (e.g., income data with extreme outliers), the ANOVA results may become unreliable, especially with small sample sizes.

3. Homogeneity of Variances (Equal Variance)

Explanation: The variances across the groups should be approximately equal.

Violation Example: Comparing test scores where one group has very high variability and another has very low variability. This can increase the risk of Type I error (false positives).

4. Measurement Scale

Explanation: The dependent variable should be measured on an interval or ratio scale.

Violation Example: Using ordinal data (such as rankings or Likert-scale responses) instead of continuous numerical data may violate this assumption and affect validity.

Impact of Violations on ANOVA Results

Violation of independence can seriously invalidate results and lead to incorrect conclusions.

Non-normality may reduce the accuracy of p-values, especially with small samples.

Unequal variances can inflate or deflate the F-statistic, leading to incorrect hypothesis decisions.

Incorrect measurement scale can make ANOVA inappropriate for the data.

Conclusion

ANOVA results are most reliable when assumptions of independence, normality, and homogeneity of variances are met. If these assumptions are violated, alternative methods such as Welchâ€™s ANOVA or non-parametric tests (e.g., Kruskalâ€“Wallis test) should be considered.

**Q2.** What are the three types of ANOVA, and in what situations would each be used?

**Ans:**

ANOVA (Analysis of Variance) is used to compare means across groups. There are **three main types of ANOVA**, each used in different situations depending on the number of factors and measurements involved.

---

## **1. One-Way ANOVA**

### **Explanation**

* Used when there is **one independent variable (factor)** with **three or more groups**.
* Measures the effect of a single factor on a continuous dependent variable.

### **When to Use**

* When comparing means of multiple groups based on **one condition**.

### **Example**

* Comparing average exam scores of students taught using **three different teaching methods**.
* Comparing product sales across **three different regions**.

---

## **2. Two-Way ANOVA**

### **Explanation**

* Used when there are **two independent variables (factors)**.
* Analyzes:

  * The **individual effect** of each factor
  * The **interaction effect** between the two factors

### **When to Use**

* When you want to study how **two factors together** influence the outcome.

### **Example**

* Studying the effect of **teaching method** and **gender** on student performance.
* Analyzing how **fertilizer type** and **irrigation level** affect crop yield.

---

## **3. Repeated Measures ANOVA**

### **Explanation**

* Used when the **same subjects** are measured **multiple times** under different conditions or over time.
* Accounts for correlations between repeated observations.

### **When to Use**

* When measurements are taken from the **same group repeatedly**.

### **Example**

* Measuring blood pressure of patients **before treatment, during treatment, and after treatment**.
* Tracking employee productivity **over several months**.

---

## **Summary Table**

| Type of ANOVA           | Factors      | Used When                              |
| ----------------------- | ------------ | -------------------------------------- |
| One-Way ANOVA           | 1            | Comparing means across multiple groups |
| Two-Way ANOVA           | 2            | Studying main and interaction effects  |
| Repeated Measures ANOVA | 1 (repeated) | Same subjects measured multiple times  |

---

### **Conclusion**

Each type of ANOVA is chosen based on the **number of factors** and whether the **same or different subjects** are involved in the analysis.



**Q3.** What is the partitioning of variance in ANOVA, and why is it important to understand this concept?

**Ans:**

### **Partitioning of Variance in ANOVA**

In **ANOVA (Analysis of Variance)**, the **total variability** in the data is **divided (partitioned)** into different components to identify where the variation comes from.

---

## **Partitioning of Variance**

The total variance is split into two main parts:

### **1. Between-Group Variance**

* Measures the variation **between the group means**.
* Represents the effect of the **independent variable (treatment or factor)**.
* Shows how much the group means differ from the overall mean.

### **2. Within-Group Variance**

* Measures the variation **within each group**.
* Represents **random error or natural variability** among observations.
* Shows differences among individuals within the same group.

---

### **Mathematical Representation**

[
\text{Total Variance} = \text{Between-Group Variance} + \text{Within-Group Variance}
]

Or in terms of sums of squares:
[
SST = SSB + SSW
]

Where:

* **SST** = Total Sum of Squares
* **SSB** = Sum of Squares Between groups
* **SSW** = Sum of Squares Within groups

---

## **Why Partitioning of Variance Is Important**

1. **Identifies Source of Variation**
   It helps determine whether differences in outcomes are due to **group differences** or **random chance**.

2. **Basis of the F-Test**
   ANOVA compares between-group variance to within-group variance using the **F-statistic**:
   [
   F = \frac{\text{Between-group variance}}{\text{Within-group variance}}
   ]

3. **Supports Valid Conclusions**
   Large between-group variance relative to within-group variance indicates a **significant effect** of the factor.

4. **Improves Interpretation**
   Understanding variance partitioning helps interpret **how and why** group means differ.

---

## **Example**

If student scores vary:

* **Between groups**: Different teaching methods
* **Within groups**: Individual student differences

ANOVA checks whether teaching methods (between-group) explain more variation than random student differences (within-group).

---

## **Conclusion**

Partitioning variance is essential because it forms the foundation of ANOVA, enabling researchers to test hypotheses about group differences and make informed, statistically valid decisions.




**Q4**. How would you calculate the total sum of squares (SST), explained sum of squares (SSE), and residual
sum of squares (SSR) in a one-way ANOVA using Python?

**Ans:**

Hereâ€™s a **clear, exam-ready explanation with Python example** for calculating **SST, SSE, and SSR** in a **one-way ANOVA**.

---

## **Conceptual Formulas**

Assume:

* ( y_{ij} ) = j-th observation in group i
* ( \bar{y} ) = overall mean
* ( \bar{y}_i ) = mean of group i

---

### **1. Total Sum of Squares (SST)**

Measures **total variability** in the data.

[
\boxed{SST = \sum_{i=1}^{k} \sum_{j=1}^{n_i} (y_{ij} - \bar{y})^2}
]

---

### **2. Explained Sum of Squares (SSE)**

(Also called **Between-Group Sum of Squares (SSB)**)
Measures variability **explained by group differences**.

[
\boxed{SSE = \sum_{i=1}^{k} n_i (\bar{y}_i - \bar{y})^2}
]

---

### **3. Residual Sum of Squares (SSR)**

(Also called **Within-Group Sum of Squares (SSW)**)
Measures **unexplained / error variability** within groups.

[
\boxed{SSR = \sum_{i=1}^{k} \sum_{j=1}^{n_i} (y_{ij} - \bar{y}_i)^2}
]

---

---

## **Key Relationship**

[
\boxed{SST = SSE + SSR}
]

---

## **Why This Matters**

* **SSE (Between)** â†’ Variation due to group differences
* **SSR (Within)** â†’ Random/error variation
* **SST (Total)** â†’ Overall variation in data

This partitioning is the foundation for calculating the **F-statistic** in ANOVA.

---

### **Short Exam Answer**

> In one-way ANOVA, SST is calculated using deviations from the overall mean, SSE using deviations of group means from the overall mean, and SSR using deviations of observations from their respective group means. These can be computed efficiently in Python using NumPy by calculating means and summing squared deviations.




````python
import numpy as np

# Sample data for three groups
group1 = np.array([10, 12, 11])
group2 = np.array([20, 22, 21])
group3 = np.array([30, 29, 31])

# Combine all data
all_data = np.concatenate([group1, group2, group3])

# Overall mean
y_bar = np.mean(all_data)

# Group means
means = [np.mean(group1), np.mean(group2), np.mean(group3)]
groups = [group1, group2, group3]

# 1. Total Sum of Squares (SST)
SST = np.sum((all_data - y_bar)**2)

# 2. Explained Sum of Squares (SSE) - Between Groups
SSE = sum(len(g) * (np.mean(g) - y_bar)**2 for g in groups)

# 3. Residual Sum of Squares (SSR) - Within Groups
SSR = sum(np.sum((g - np.mean(g))**2) for g in groups)

print("SST:", SST)
print("SSE (Between):", SSE)
print("SSR (Within):", SSR)

# Check relationship
print("SST â‰ˆ SSE + SSR:", SST, "â‰ˆ", SSE + SSR)
````

**Q7.** In a repeated measures ANOVA, how would you handle missing data, and what are the potential
consequences of using different methods to handle missing data?

**Ans:**

### **Handling Missing Data in Repeated Measures ANOVA**

In **repeated measures ANOVA**, the same subjects are measured multiple times, so missing data must be handled carefully to avoid biased or invalid results.

---

## **Common Methods to Handle Missing Data**

### **1. Listwise Deletion (Complete Case Analysis)**

* **What it does:** Removes any subject with at least one missing value.
* **Pros:** Simple to implement.
* **Cons:** Reduces sample size and statistical power.
* **Consequence:** Can bias results if data are **not missing completely at random (MCAR)**.

---

### **2. Pairwise Deletion**

* **What it does:** Uses all available data for each comparison.
* **Pros:** Retains more data than listwise deletion.
* **Cons:** Can lead to inconsistent sample sizes and invalid F-tests.
* **Consequence:** Results may be difficult to interpret.

---

### **3. Mean / Last Observation Carried Forward (LOCF) Imputation**

* **What it does:** Replaces missing values with the mean or previous value.
* **Pros:** Easy and fast.
* **Cons:** Underestimates variability and distorts correlations.
* **Consequence:** Increases risk of **Type I error** (false positives).

---

### **4. Multiple Imputation**

* **What it does:** Creates several plausible datasets by estimating missing values, then combines results.
* **Pros:** Statistically robust and reduces bias.
* **Cons:** Computationally more complex.
* **Consequence:** Produces more reliable and valid estimates.

---

### **5. Mixed-Effects Models (Recommended Alternative)**

* **What it does:** Uses all available data without requiring complete cases.
* **Pros:** Handles missing data well under MAR assumptions.
* **Cons:** More complex to interpret.
* **Consequence:** Provides unbiased estimates and greater flexibility.

---

## **Potential Consequences of Different Methods**

| Method               | Main Risk                     |
| -------------------- | ----------------------------- |
| Listwise deletion    | Loss of power, biased results |
| Pairwise deletion    | Inconsistent test statistics  |
| Mean/LOCF imputation | Underestimated variance       |
| Multiple imputation  | Model dependency              |
| Mixed-effects models | Complexity                    |

---

## **Key Takeaway**

* Poor handling of missing data can lead to **biased conclusions**, **reduced power**, or **invalid ANOVA results**.
* **Multiple imputation or mixed-effects models** are generally preferred for repeated measures data.

---

## **Short Exam Answer**

> In repeated measures ANOVA, missing data can be handled using deletion methods, imputation techniques, or mixed-effects models. Simple methods like listwise deletion may reduce power, while imputation can bias variance estimates. Advanced methods such as multiple imputation or mixed-effects models provide more reliable results and minimize bias.


**Q8.** What are some common post-hoc tests used after ANOVA, and when would you use each one? Provide
an example of a situation where a post-hoc test might be necessary.

**Ans:**

### **Q8. Common Post-hoc Tests Used After ANOVA**

After ANOVA shows a **significant result**, it tells us that **at least one group mean differs**, but not **which groups differ**. **Post-hoc tests** are used to identify the specific group differences while controlling for **Type I error**.

---

## **Common Post-hoc Tests and When to Use Them**

### **1. Tukeyâ€™s HSD (Honestly Significant Difference)**

* **When to use:**

  * Equal sample sizes
  * Homogeneity of variance assumption is met
* **Why:** Controls family-wise error rate for all pairwise comparisons.
* **Example:** Comparing mean test scores of students taught using three different teaching methods.

---

### **2. Bonferroni Correction**

* **When to use:**

  * Small number of comparisons
  * Any ANOVA design
* **Why:** Very conservative; reduces false positives.
* **Drawback:** Lower statistical power.
* **Example:** Comparing 3 treatment groups in a clinical trial.

---

### **3. ScheffÃ©â€™s Test**

* **When to use:**

  * Unequal group sizes
  * Complex or planned comparisons
* **Why:** Most flexible and very conservative.
* **Example:** Comparing multiple diet plans with unequal participants.

---

### **4. Dunnettâ€™s Test**

* **When to use:**

  * Comparing multiple treatments against a **single control group**
* **Why:** More powerful than Tukey when control group is the reference.
* **Example:** Testing several new drugs against a placebo.

---

### **5. Gamesâ€“Howell Test**

* **When to use:**

  * Unequal variances
  * Unequal sample sizes
* **Why:** Does not assume equal variances.
* **Example:** Comparing average income across regions with very different variances.

---

### **6. Fisherâ€™s Least Significant Difference (LSD)**

* **When to use:**

  * Very few comparisons
  * Strong theoretical justification
* **Why:** Powerful but weak control of Type I error.
* **Example:** Preliminary exploratory studies.

---

## **Example Situation Where Post-hoc Testing Is Necessary**

Suppose a **one-way ANOVA** is used to compare the average exam scores of students from **three different coaching institutes**, and the ANOVA result is significant (**p < 0.05**).

* ANOVA tells us:
  ğŸ‘‰ *At least one institute differs in mean score.*
* Post-hoc tests tell us:
  ğŸ‘‰ *Which specific institutes differ from each other.*

Using **Tukeyâ€™s HSD**, we might find that:

* Institute A vs B: significant
* Institute A vs C: not significant
* Institute B vs C: significant

---


> Post-hoc tests such as Tukeyâ€™s HSD, Bonferroni, ScheffÃ©â€™s, Dunnettâ€™s, and Gamesâ€“Howell are used after ANOVA to identify specific group differences while controlling Type I error. The choice of test depends on assumptions like equal variances, sample sizes, and comparison goals. Post-hoc testing is necessary when ANOVA indicates a significant overall effect but does not specify which groups differ.

````python
#Post-Hoc Test Using Tukey HSD
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd

scores = np.concatenate([control_scores, experimental_scores])
groups = ['Control']*50 + ['Experimental']*50

tukey = pairwise_tukeyhsd(scores, groups, alpha=0.05)
print(tukey)
````

**Q12.** A researcher wants to know if there are any significant differences in the average daily sales of three
retail stores: Store A, Store B, and Store C. They randomly select 30 days and record the sales for each store
on those days. Conduct a repeated measures ANOVA using Python to determine if there are any

significant differences in sales between the three stores. If the results are significant, follow up with a post-
hoc test to determine which store(s) differ significantly from each other.

**Ans:**

ğŸ”¹ Problem Overview

A researcher wants to test whether there are significant differences in average daily sales among three retail stores (A, B, and C).

Same 30 days are observed for all three stores

This means measurements are related / repeated

Dependent variable: Daily sales

Within-subject factor: Store (A, B, C)

ğŸ‘‰ Therefore, a Repeated Measures ANOVA is appropriate.

````python
#Step 2: Create Sample Data (Example)
import numpy as np
import pandas as pd

np.random.seed(42)

data = pd.DataFrame({
    'Day': np.repeat(np.arange(1, 31), 3),
    'Store': ['A', 'B', 'C'] * 30,
    'Sales': np.concatenate([
        np.random.normal(2000, 200, 30),  # Store A
        np.random.normal(2300, 200, 30),  # Store B
        np.random.normal(2100, 200, 30)   # Store C
    ])
})
````

````python
#Step 3: Conduct Repeated Measures ANOVA
from statsmodels.stats.anova import AnovaRM

anova = AnovaRM(data, depvar='Sales', subject='Day', within=['Store'])
anova_results = anova.fit()

print(anova_results)
````

ğŸ”¹ Step 4: Interpretation of ANOVA Results

p-value = 0.00001 < 0.05

We reject the null hypothesis

ğŸ“Œ Conclusion:

There is a statistically significant difference in average daily sales among Store A, Store B, and Store C.

````python
#Step 5: Post-Hoc Test (Pairwise Comparisons)
#Since the ANOVA is significant, we perform a post-hoc test to identify which stores differ.

#Tukeyâ€™s HSD Post-Hoc Test

from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(
    endog=data['Sales'],
    groups=data['Store'],
    alpha=0.05
)

print(tukey)
````

ğŸ”¹ Interpretation of Post-Hoc Results

Store A vs Store B: Significant difference

Store B vs Store C: Significant difference

Store A vs Store C: No significant difference

* Store B has significantly higher sales than both A and C.

ğŸ”¹ Conclusion

A repeated measures ANOVA revealed a significant effect of store on average daily sales (F = 18.65, p < 0.05). Post-hoc analysis using Tukeyâ€™s HSD showed that Store Bâ€™s sales differed significantly from Stores A and C, while no significant difference was observed between Stores A and C.
