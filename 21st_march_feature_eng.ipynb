**Q1.** What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you
might choose one over the other.

**Ans:**
## **1. Definition**

| Encoding Type        | Description                                                                    |
| -------------------- | ------------------------------------------------------------------------------ |
| **Label Encoding**   | Assigns an **integer value** to each category **without implying any order**.  |
| **Ordinal Encoding** | Assigns an **integer value** to each category **based on a meaningful order**. |

---

## **2. Key Differences**

| Feature              | Label Encoding                             | Ordinal Encoding                        |
| -------------------- | ------------------------------------------ | --------------------------------------- |
| Order implied        | ❌ No order implied                         | ✅ Preserves meaningful order            |
| Suitable for         | Nominal categorical variables              | Ordinal categorical variables           |
| Risk                 | May mislead algorithms if order is assumed | Correctly represents ordered categories |
| Number of categories | Any number                                 | Any number                              |

---

## **3. Example**

**Dataset Example:** Education level of employees

| Education Level | Integer Mapping (Ordinal) |
| --------------- | ------------------------- |
| High School     | 1                         |
| Bachelor's      | 2                         |
| Master's        | 3                         |
| PhD             | 4                         |

* **Ordinal Encoding:** Preserves order → Master’s > Bachelor’s
* **Label Encoding:** If you use Label Encoding arbitrarily → `High School=0, Bachelor's=1, Master's=2, PhD=3`, ML algorithm might **misinterpret distances**, which is fine only for tree-based models, but can mislead linear models.

---

**Nominal Example (No order): Gender**

* Categories: `Male`, `Female`, `Other`
* **Label Encoding:** Male=0, Female=1, Other=2 → suitable because **no order exists**
* **Ordinal Encoding:** ❌ Not appropriate because gender has no meaningful order

---

## **4. When to Choose Which**

1. **Ordinal Encoding:**

   * Use when categories have a **natural ranking** (e.g., education level, customer satisfaction rating: low/medium/high)

2. **Label Encoding:**

   * Use for **nominal data without order** (e.g., country, product type, gender)

---

✅ **Summary:**

* **Ordinal Encoding → for ordered categories**
* **Label Encoding → for unordered categories**


**Q2.** Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in
a machine learning project.

**Ans:**

### **Target Guided Ordinal Encoding**

Target Guided Ordinal Encoding is a technique used to convert categorical variables into numerical values **based on their relationship with the target variable**.
Instead of assigning random numbers (like Label Encoding), this method assigns ranks according to the **mean of the target variable for each category**.

---

## **How It Works – Step by Step**

1. **Group the categorical feature by each category.**
2. **Calculate the mean (or median) of the target variable for each category.**
3. **Sort the categories based on this target mean.**
4. **Assign ordinal numbers according to this order.**

   * Higher target mean → higher rank
   * Lower target mean → lower rank

This encoding preserves the relationship between the feature and the target, which can improve model performance.

---

## **Example**

### Scenario

You are building a model to predict whether a customer will buy a product (`Purchase = 1 or 0`).
You have a categorical column **City**:

| City   | Purchase |
| ------ | -------- |
| Mumbai | 1        |
| Pune   | 0        |
| Mumbai | 1        |
| Delhi  | 0        |
| Pune   | 1        |
| Delhi  | 0        |

### Step 1: Calculate Mean Purchase per City

| City   | Mean Purchase |
| ------ | ------------- |
| Mumbai | 1.0           |
| Pune   | 0.5           |
| Delhi  | 0.0           |

### Step 2: Assign Ranks

| City   | Encoded Value |
| ------ | ------------- |
| Mumbai | 3             |
| Pune   | 2             |
| Delhi  | 1             |

Now the model understands:
👉 Mumbai customers purchase more → highest value.

---

## **When to Use It**

✔ When the categorical feature has a **strong relationship with the target**
✔ Useful in:

* Customer churn prediction
* Credit risk modeling
* Sales prediction
* Loan approval systems

### Example Use Case

In a **loan default prediction project**, the column “Occupation” may influence default rate:

| Occupation | Default Rate |
| ---------- | ------------ |
| Student    | 0.40         |
| Business   | 0.25         |
| Govt Job   | 0.10         |

Target guided encoding will assign:

Student → 3
Business → 2
Govt Job → 1

This helps the model learn real patterns.

---

## **Advantages**

* Captures relationship with target
* More meaningful than label encoding
* Works well with tree-based models

## **Disadvantages**

* Risk of **data leakage** if applied before train-test split
* Can cause overfitting on small datasets




**Q3.** Define covariance and explain why it is important in statistical analysis. How is covariance calculated?

**Ans:**

### **Q3. Define Covariance and Explain Its Importance. How Is Covariance Calculated?**

---

## **1. What is Covariance?**

**Covariance** is a statistical measure that describes the **direction of the relationship** between two variables.

It tells us:

* Whether two variables increase together → **positive covariance**
* Whether one increases while the other decreases → **negative covariance**
* Whether there is no clear relationship → **covariance near zero**

👉 Covariance does **not** measure the strength, only the **direction** of the relationship.

---

## **2. Importance of Covariance in Statistical Analysis**

Covariance is important because:

1. **Identifies Relationships Between Variables**

   * Helps understand how two features move together.

2. **Foundation of Correlation**

   * Correlation is derived from covariance (normalized version).

3. **Used in Machine Learning & Finance**

   * Feature selection
   * Portfolio risk analysis
   * PCA (Principal Component Analysis)

4. **Helps in Multivariate Analysis**

   * Determines dependency between variables in datasets.

---

## **3. Interpretation of Covariance**

| Covariance Value | Meaning                                      |
| ---------------- | -------------------------------------------- |
| Positive (+)     | Both variables increase or decrease together |
| Negative (–)     | One increases, the other decreases           |
| Zero             | No linear relationship                       |

---

## **4. Formula for Covariance**

For two variables X and Y with n observations:

### **Population Covariance**

[
Cov(X,Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{n}
]

### **Sample Covariance**

[
Cov(X,Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{n-1}
]

Where:

* (X_i, Y_i) = individual values
* (\bar{X}, \bar{Y}) = means of X and Y
* n = number of observations

---

## **5. Simple Example**

Suppose:

X = [2, 4, 6]
Y = [3, 7, 9]

1. Mean of X = 4
2. Mean of Y = 6.33

Compute:

[
Cov(X,Y) = \frac{(2-4)(3-6.33) + (4-4)(7-6.33) + (6-4)(9-6.33)}{3-1}
]

Covariance will be **positive**, meaning X and Y move in same direction.

---

## **6. Limitation of Covariance**

❌ Value depends on units (not standardized)
❌ Hard to compare across datasets

👉 That’s why we use **Correlation**, which is normalized between -1 and +1.

**Q4.** For a dataset with the following categorical variables: Color (red, green, blue), Size (small, medium,
large), and Material (wood, metal, plastic), perform label encoding using Python's scikit-learn library.
Show your code and explain the output.

**Ans:**

Step 1 – Create the Dataset

We have 3 categorical features:

Color: red, green, blue

Size: small, medium, large

Material: wood, metal, plastic

Explanation
1. What LabelEncoder Does

Converts each category → numeric value

Encoding is based on alphabetical order

Example mapping:

Color

blue → 0

green → 1

red → 2

Size

large → 0

medium → 1

small → 2

Material

metal → 0

plastic → 1

wood → 2

Why Use Label Encoding?

✅ Converts text → numbers for ML models
✅ Simple and memory efficient
✅ Useful when:

Categorical variable has ordinal meaning

Algorithms like Decision Tree, Random Forest

❌ Not ideal for nominal data with no order (use One-Hot Encoding instead)

Conclusion

Label Encoding assigns an integer to each unique category.

scikit-learn’s LabelEncoder() is used feature-wise.

Output shows numeric representation of Color, Size, Material.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create dataset
data = pd.DataFrame({
    'Color': ['red', 'green', 'blue', 'green', 'red'],
    'Size': ['small', 'medium', 'large', 'medium', 'small'],
    'Material': ['wood', 'metal', 'plastic', 'wood', 'metal']
})

print("Original Data:")
print(data)

# Apply Label Encoding
encoder = LabelEncoder()

data['Color_encoded'] = encoder.fit_transform(data['Color'])
data['Size_encoded'] = encoder.fit_transform(data['Size'])
data['Material_encoded'] = encoder.fit_transform(data['Material'])

print("\nAfter Label Encoding:")
print(data)
```

**Q6.** You are working on a machine learning project with a dataset containing several categorical
variables, including "Gender" (Male/Female), "Education Level" (High School/Bachelor's/Master's/PhD),
and "Employment Status" (Unemployed/Part-Time/Full-Time). Which encoding method would you use for
each variable, and why?

**Ans:**


We have three categorical features:

1. **Gender** → Male / Female
2. **Education Level** → High School / Bachelor’s / Master’s / PhD
3. **Employment Status** → Unemployed / Part-Time / Full-Time

Each variable has a different nature, so we choose different encoding techniques.

---

## **1. Gender (Male / Female)**

### ✅ Recommended: **Label Encoding or Binary Encoding**

* Gender has only **two categories** → binary nature.
* No risk of creating false ordering (0 and 1 simply represent two groups).
* One-hot encoding would create two columns unnecessarily.

**Example Encoding**

| Gender | Encoded |
| ------ | ------- |
| Male   | 0       |
| Female | 1       |

**Reason**

* Simple and efficient
* Works well with most ML algorithms
* No ordinal meaning is implied between Male and Female.

---

## **2. Education Level (High School < Bachelor < Master < PhD)**

### ✅ Recommended: **Ordinal Encoding**

* Education has a **natural order / ranking**
* Higher education usually indicates higher skill level.

**Logical Order**

| Education Level | Code |
| --------------- | ---- |
| High School     | 0    |
| Bachelor        | 1    |
| Master          | 2    |
| PhD             | 3    |

**Reason**

* Preserves meaningful hierarchy
* Models like Linear Regression, Decision Trees can use this order
* One-hot would destroy ordinal information.

---

## **3. Employment Status (Unemployed / Part-Time / Full-Time)**

### ✅ Recommended: **One-Hot Encoding**

* Categories are **nominal** (no true mathematical order).
* Full-Time is not “greater” than Part-Time in a numeric sense.

**One-Hot Result**

| Unemployed | Part-Time | Full-Time |
| ---------- | --------- | --------- |
| 1          | 0         | 0         |
| 0          | 1         | 0         |
| 0          | 0         | 1         |

**Reason**

* Prevents model from assuming false ranking
* Suitable for algorithms like:

  * Logistic Regression
  * SVM
  * Neural Networks

---

# **Final Summary Table**

| Feature           | Type           | Best Encoding    | Reason                                     |
| ----------------- | -------------- | ---------------- | ------------------------------------------ |
| Gender            | Binary Nominal | Label Encoding   | Only two categories, simple representation |
| Education Level   | Ordinal        | Ordinal Encoding | Has natural ranking                        |
| Employment Status | Nominal        | One-Hot Encoding | No inherent order                          |

---

## **Conclusion**

* ✅ Use encoding based on **nature of data**, not randomly
* Preserve order when it exists
* Avoid creating false relationships

This improves:

* Model accuracy
* Interpretability
* Training efficiency
