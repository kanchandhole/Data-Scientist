**Q1.** Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.

**Ans:**

Sure! Let's solve this **step by step** using Python.

We are asked to **calculate a 95% confidence interval** for:

* Sample mean ((\bar{x})) = 50
* Sample standard deviation ((s)) = 5
* Sample size ((n)) → let's assume 30 (if not specified)
* Confidence level = 95%

---

### **Step 1: Formula for confidence interval (t-distribution since σ unknown)**

[
CI = \bar{x} \pm t_{\alpha/2, , df} \cdot \frac{s}{\sqrt{n}}
]

Where:

* (t_{\alpha/2, , df}) = t critical value for confidence level
* (s/\sqrt{n}) = standard error
* df = n - 1

---

### **Step 2: Python Implementation**

```python

```

---

### **Step 3: Step-by-step Calculation**

1. Standard error:
   [
   SE = \frac{s}{\sqrt{n}} = \frac{5}{\sqrt{30}} \approx 0.913
   ]

2. t critical value (df = 29, 95% CI):
   [
   t_{0.975,29} \approx 2.045
   ]

3. Margin of error:
   [
   ME = 2.045 \times 0.913 \approx 1.867
   ]

4. Confidence interval:
   [
   CI = 50 \pm 1.867 = [48.13, 51.87]
   ]

---

### **Step 4: Interpretation**

We are **95% confident** that the true population mean lies between **48.13 and 51.87**.



````python
import scipy.stats as stats
import math

# Given data
sample_mean = 50
sample_std = 5
n = 30
confidence = 0.95

# Degrees of freedom
df = n - 1

# Standard error
SE = sample_std / math.sqrt(n)

# t critical value
t_critical = stats.t.ppf((1 + confidence)/2, df)

# Margin of error
ME = t_critical * SE

# Confidence interval
CI_lower = sample_mean - ME
CI_upper = sample_mean + ME

print(f"95% Confidence Interval: ({CI_lower:.2f}, {CI_upper:.2f})")
````

````python
import scipy.stats as stats
import numpy as np

# Observed counts
observed = np.array([21, 19, 18, 10, 9, 23])

# Expected percentages
expected_percent = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2])

# Total number of M&Ms
n_total = observed.sum()

# Expected counts
expected = expected_percent * n_total

# Perform chi-square test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Distribution does NOT match expected proportions")
else:
    print("Fail to reject null hypothesis: Distribution matches expected proportions")
````

**Q4.** A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.

**Ans:**

Sure! Let's solve this **step by step** using Python.

We are calculating a **95% confidence interval (CI) for a population proportion**.

---

## **Step 1: Define the problem**

* Total population (sample size): (n = 500)
* Number of smokers: (x = 60)
* Sample proportion:
  [
  \hat{p} = \frac{x}{n} = \frac{60}{500} = 0.12
  ]
* Confidence level = 95% → z = 1.96

---

## **Step 2: Formula for CI for proportion**

[
CI = \hat{p} \pm z \cdot \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}
]

Where:

* (\hat{p}) = sample proportion
* (n) = sample size
* (z) = z-score for confidence level

---

## **Step 3: Python Implementation**

```python
import math
import scipy.stats as stats

# Given data
n = 500
x = 60
p_hat = x / n
confidence = 0.95

# Standard error
SE = math.sqrt(p_hat * (1 - p_hat) / n)

# z critical value
z_critical = stats.norm.ppf((1 + confidence) / 2)

# Margin of Error
ME = z_critical * SE

# Confidence Interval
CI_lower = p_hat - ME
CI_upper = p_hat + ME

print(f"Sample proportion: {p_hat:.3f}")
print(f"95% Confidence Interval: ({CI_lower:.3f}, {CI_upper:.3f})")
```

---

## **Step 4: Step-by-step calculation**

1. **Sample proportion:**
   [
   \hat{p} = \frac{60}{500} = 0.12
   ]

2. **Standard error (SE):**
   [
   SE = \sqrt{\frac{0.12 \cdot (1-0.12)}{500}} = \sqrt{\frac{0.1056}{500}} \approx \sqrt{0.0002112} \approx 0.01453
   ]

3. **Margin of Error (ME):**
   [
   ME = z \cdot SE = 1.96 \cdot 0.01453 \approx 0.0285
   ]

4. **Confidence Interval:**
   [
   CI = 0.12 \pm 0.0285
   ]
   [
   \text{Lower bound} = 0.12 - 0.0285 \approx 0.0915
   ]
   [
   \text{Upper bound} = 0.12 + 0.0285 \approx 0.1485
   ]

---

## **Step 5: Interpretation**

We are **95% confident** that the true proportion of smokers in the population is between **9.2% and 14.9%**.




````python
## **Step 3: Python Implementation**


import math
import scipy.stats as stats

# Given data
n = 500
x = 60
p_hat = x / n
confidence = 0.95

# Standard error
SE = math.sqrt(p_hat * (1 - p_hat) / n)

# z critical value
z_critical = stats.norm.ppf((1 + confidence) / 2)

# Margin of Error
ME = z_critical * SE

# Confidence Interval
CI_lower = p_hat - ME
CI_upper = p_hat + ME

print(f"Sample proportion: {p_hat:.3f}")
print(f"95% Confidence Interval: ({CI_lower:.3f}, {CI_upper:.3f})")
````

* Confidence level = 90%

We **assume the sample size is large** or population is approximately normal. If sample size isn’t given, we’ll assume a **z-distribution**.

The **formula for confidence interval** is:

[
CI = \bar{x} \pm z_{\alpha/2} \times \frac{s}{\sqrt{n}}

Where:

* (z_{\alpha/2}) is the z-score for the confidence level (for 90%, (z_{\alpha/2} \approx 1.645))

**Q6.** Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.

**Ans:**  df = 10 → degrees of freedom.

chi2.pdf(x, df) → calculates the probability density function of the chi-square distribution.

The shaded region represents the area to the right of χ² = 15, which corresponds to the probability of getting a chi-square statistic greater than 15.

This plot visually shows how extreme the value 15 is relative to a chi-square distribution with 10 degrees of freedom.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Given data
df = 10  # degrees of freedom
chi_square_value = 15

# Create an array of x values (chi-square values)
x = np.linspace(0, 30, 1000)

# Compute the chi-square PDF
y = chi2.pdf(x, df)

# Plot the chi-square distribution
plt.figure(figsize=(8,5))
plt.plot(x, y, label=f'Chi-square PDF (df={df})', color='blue')

# Shade the area corresponding to chi-square >= 15
x_shade = np.linspace(chi_square_value, 30, 500)
y_shade = chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, color='orange', alpha=0.5, label=f'Chi-square >= {chi_square_value}')

# Label axes and add title
plt.xlabel('Chi-square value')
plt.ylabel('Probability Density')
plt.title('Chi-square Distribution')
plt.legend()
plt.grid(True)
plt.show()
````

### **Step 1: Define the hypotheses**

* **Alternative hypothesis ((H_a))**: The coin is biased (probabilities are not 0.5)

This is a **two-sided test**.

---

### **Step 2: Observed and expected frequencies**

* Number of flips: ( n = 100 )

* Observed tails = 45 → observed heads = 55

  [
  E = n \times 0.5 = 100 \times 0.5 = 50

  So, ([50, 50])

### **Step 3: Chi-square test statistic**

[
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
]

Where:

* (O_i) = observed frequency
* (E_i) = expected frequency

---

### **Step 4: Python Code**

```python

```

---

### **Step 5: Manual Calculation (Optional)**

[
\chi^2 = \frac{(55-50)^2}{50} + \frac{(45-50)^2}{50} = \frac{25}{50} + \frac{25}{50} = 0.5 + 0.5 = 1.0
]

Degrees of freedom: ( df = k-1 = 2-1 = 1 )

P-value for (\chi^2 = 1) and (df=1) ≈ 0.317

---

### **Step 6: Interpretation**

* **Chi-square statistic**: 1.0
* **P-value**: 0.317 (> 0.05)

````python
import numpy as np
from scipy.stats import chi2_contingency

# Observed frequencies
observed = np.array([[200, 150, 150],
                     [225, 175, 100]])

# Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(observed)

print("Chi-square statistic:", round(chi2_stat, 4))
print("P-value:", round(p_value, 4))
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)

# Conclusion
alpha = 0.01
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between chocolate preference and country.")
else:
    print("Fail to reject the null hypothesis: No significant association between chocolate preference and country.")
````

**Q11.** A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.

**Ans:**

Sure! Let's carefully solve this step by step.

We are asked to **conduct a hypothesis test** when the population mean and standard deviation are unknown. Since the sample size is **n = 30**, and population standard deviation is unknown, we will use a **t-test**.

---

### **Given:**

* Sample size: ( n = 30 )
* Sample mean: ( \bar{x} = 72 )
* Sample standard deviation: ( s = 10 )
* Population mean under null hypothesis: ( \mu_0 = 70 )
* Significance level: ( \alpha = 0.05 )

---

### **Step 1: State the hypotheses**

* **Null hypothesis (H₀):** ( \mu = 70 )
* **Alternative hypothesis (H₁):** ( \mu \neq 70 ) (two-tailed test)

---

### **Step 2: Test statistic**

For a **t-test**, the test statistic is:

[
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
]

Substitute the values:

[
t = \frac{72 - 70}{10 / \sqrt{30}}
]

First, calculate the **standard error (SE):**

[
SE = \frac{s}{\sqrt{n}} = \frac{10}{\sqrt{30}}
]

(\sqrt{30} \approx 5.477)

[
SE \approx \frac{10}{5.477} \approx 1.825
]

Now, calculate t:

[
t = \frac{2}{1.825} \approx 1.096
]

---

### **Step 3: Degrees of freedom**

[
df = n - 1 = 30 - 1 = 29
]

---

### **Step 4: Critical t-value**

For a **two-tailed test** at ( \alpha = 0.05 ) and df = 29:

[
t_{critical} \approx \pm 2.045
]

---

### **Step 5: Compare t-statistic with t-critical**

* Calculated t = 1.096
* Critical t = ±2.045

Since ( |1.096| < 2.045 ), **we fail to reject the null hypothesis**.

---

### **Step 6: Conclusion**

At the 0.05 significance level, there is **not enough evidence** to conclude that the population mean is significantly different from 70.
