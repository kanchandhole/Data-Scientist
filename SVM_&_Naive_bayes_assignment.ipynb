**Q2.** What is the difference between Hard Margin and Soft Margin SVM?

**ans:**The main difference between hard margin and soft margin SVMs lies in their approach to handling misclassifications. Hard margin SVMs aim to find a hyperplane that perfectly separates the data into classes without any misclassifications, while soft margin SVMs allow for some misclassifications to achieve better generalization and avoid overfitting.

**Q3.**What is the mathematical intuition behind SVM.

**ans:** key mathematical concepts:
1. Hyperplane and Margin:
SVMs aim to find a hyperplane, which is a decision boundary that separates different classes.
The "margin" is the distance between the hyperplane and the closest data points on either side.
The goal is to maximize this margin.
2. Support Vectors:
Support vectors are the data points that lie closest to the hyperplane and are crucial in determining its position and orientation.
Changing the position of these support vectors will change the hyperplane.
3. Linear Separability:
SVMs are initially designed for linearly separable data, where a hyperplane can perfectly separate the classes.
In practice, data may not be perfectly separable, so techniques like soft margins are used to allow for some misclassification.
4. Optimization:
The process of finding the optimal hyperplane involves solving a quadratic programming problem.
This optimization aims to minimize a specific cost function while satisfying certain constraints.
5. Kernel Functions:
Kernel functions are used to map data into higher-dimensional spaces, making it easier to find a linear separation even when the original data is not linearly separable.
Common kernel functions include polynomial, radial basis function (RBF), and sigmoid kernels.
6. Soft Margin:
Soft margin SVMs allow for some misclassification by introducing slack variables (Î¾) that penalize misclassified data points.
This makes the algorithm more flexible and robust to noisy data.
7. Dual Representation:
SVMs can be formulated in both primal and dual forms.
The dual form often simplifies the optimization process, especially for large datasets.

**Q6.** What is a Support Vector Classifier (SVC)?

**ans:** A Support Vector Classifier (SVC) is a type of Support Vector Machine (SVM) specifically designed for classification tasks. It aims to find the optimal hyperplane that best separates different classes of data points, maximizing the margin between the hyperplane and the nearest data points (support vectors).

What it does:

SVC uses a hyperplane to classify data points into different categories. It finds the hyperplane that provides the largest separation (margin) between classes, making it a robust classifier.

How it works:

Hyperplane: SVC defines a hyperplane, which is a separating boundary in a multi-dimensional space. In 2D, it's a line, and in 3D, it's a plane.
Margin: The margin is the distance between the hyperplane and the nearest data points (support vectors) from each class. A larger margin indicates a more

robust and reliable classifier.
Support Vectors: These are the data points closest to the hyperplane. They play a crucial role in defining the position of the hyperplane.

Key Concepts:

Binary Classification: SVC is often used for binary classification (two classes) but can be extended to multi-class scenarios.
Kernel Trick: SVC can use kernel functions to map data into higher-dimensional spaces, allowing for non-linear decision boundaries.

Linear vs. Non-linear: If the data is linearly separable, SVC will find a linear hyperplane. For non-linear data, it uses kernel functions to find a non-linear boundary.

When to use SVC:
Classification Problems: SVC is well-suited for classification tasks where you want to predict the class of new data points.
High-Dimensional Data: SVC can handle high-dimensional data effectively.
Non-linear Data: Using kernel functions, SVC can handle non-linear relationships in the data.

Advantages of SVC:
Effective in High-Dimensional Spaces: SVC can handle data with many features.
Relatively Robust to Overfitting: SVC tends to be less prone to overfitting than some other algorithms.
Effective with Non-linear Data: The kernel trick allows SVC to find complex decision boundaries.

Disadvantages of SVC:
Computational Cost: Training SVC can be computationally expensive, especially with large datasets.
Parameter Tuning: Finding the right parameters for SVC (like the kernel and C parameter) can be challenging.

For complex data, polynomial or RBF kernels are more suitable.
Consider computational resources when choosing between kernels.
RBF kernels are a good default choice for non-linear problems, but may require careful tuning of the gamma parameter.

**Q10.**What is the effect of the C parameter in SVM?

**ans:** In Support Vector Machines (SVM), the 'C' parameter, also known as the regularization parameter, controls the trade-off between maximizing the margin of the decision boundary and penalizing misclassifications of training data

Large C:
The SVM model will prioritize minimizing misclassifications on the training data, even if it means accepting a smaller margin. This can lead to a model that fits the training data more closely but might be more sensitive to variations in new, unseen data (overfitting).

Small C:
The SVM model will focus on maximizing the margin, allowing for a larger margin but potentially accepting more misclassifications on the training data. This can lead to a model that is more robust to new data but might not fit the training data as well (underfitting).

**Q11**.What is the role of the Gamma parameter in RBF Kernel SVM?

**ans:**In RBF kernel SVM, the gamma parameter controls the influence of a single training example on the decision boundary. It essentially determines how far the influence of a support vector extends in the higher-dimensional space created by the RBF kernel. A high gamma value makes the model sensitive to individual data points, potentially leading to overfitting, while a low gamma value results in a smoother decision boundary, potentially underfitting.

**Q12.** What is the NaÃ¯ve Bayes classifier, and why is it called "NaÃ¯ve"?

**ans:** The Naive Bayes classifier is a probabilistic classification algorithm based on Bayes' Theorem. It's called "naive" because it makes the simplifying assumption that features are conditionally independent given the class label. This means it assumes that each feature contributes to the classification decision independently of other features.

Bayes' Theorem:
Bayes' Theorem is a mathematical formula that describes the probability of an event, given prior knowledge of conditions related to the event.
Conditional Independence:
The "naive" assumption is that the presence or absence of one feature does not affect the presence or absence of any other feature. This is often not true in real-world data, hence the term "naive".

Classification:
The algorithm uses the probabilities of features to predict the class label of a new data point. It assigns the data point to the class with the highest probability.

Example:
Consider a dataset describing fruits. A naive Bayes classifier might assume that the color, shape, and size of a fruit are independent of each other when deciding if it's an apple. This means that a red, round fruit of a certain size is classified as an apple regardless of any correlation between color, shape, and size.

**Q14.**Explain the differences between Gaussian NaÃ¯ve Bayes, Multinomial NaÃ¯ve Bayes, and Bernoulli NaÃ¯ve Bayes:

**ans:** The main difference between Gaussian, Multinomial, and Bernoulli Naive Bayes lies in the probability distribution they assume for the features: Gaussian for continuous data, Multinomial for discrete counts (like word frequencies), and Bernoulli for binary or boolean features.

Gaussian Naive Bayes:
This variant assumes that features are normally distributed (Gaussian). It's suitable for tasks involving continuous data where features can take any value within a range. For example, predicting a student's height based on their weight.

Multinomial Naive Bayes:
This version assumes that features are discrete counts, often representing the frequency of occurrences (like word counts in text). It's commonly used in natural language processing (NLP) for text classification tasks. For example, classifying a student as "clever" based on their marks in various subjects.

Bernoulli Naive Bayes:
This classifier is designed for binary or boolean features, meaning each feature is either present (1) or absent (0). It's useful when the data represents the presence or absence of certain characteristics or events. For example, predicting whether a person is "healthy" based on whether they have diabetes, high blood pressure, or thyroid issues.

```python
# 4. Train the classifier
text_clf.fit(X_train, y_train)
```

```python

# 5. Predict on the test set
y_pred = text_clf.predict(X_test)
```

```python
# 6. Evaluate performance
accuracy_score(y_test, y_pred)
```

**Q29.** Write a Python program to apply feature scaling before training an SVM model and compare results with
unscaled data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
# 1. Generate synthetic dataset
X, y = make_classification(n_samples=300, n_features=2, n_redundant=0,
                           n_informative=2, random_state=42, class_sep=1.5)
```

```python
# 2. Split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# 3. Scale features using StandardScaler (fit only on train)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# 4. Train two SVM models: unscaled vs scaled
svm_unscaled = svm.SVC(kernel='rbf', gamma='scale', C=1.0)
svm_unscaled.fit(X_train, y_train)
```

**Q45.**Write a Python program to train a NaÃ¯ve Bayes classifier and evaluate its performance using the ROC-AUC
score.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_curve, auc, roc_auc_score
```

```python
# 1. Generate a binary classification dataset
X, y = make_classification(
    n_samples=1000, n_features=10, n_informative=5,
    n_redundant=0, flip_y=0.05, weights=[0.6, 0.4],
    random_state=42
)
```

```python
# 2. Split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
```

```python
# 3. Train Gaussian NaÃ¯ve Bayes
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python

# 4. Predict probabilities
y_proba = clf.predict_proba(X_test)[:, 1]
```

```python
# 5. Compute ROC metrics
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)
roc_auc_sklearn = roc_auc_score(y_test, y_proba)

print(f"ROC AUC (auc function):        {roc_auc:.4f}")
print(f"ROC AUC (roc_auc_score):      {roc_auc_sklearn:.4f}")
```

```python
# 6. Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue',
         label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve â Gaussian NaÃ¯ve Bayes")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

**Q46.**Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, average_precision_score
```

```python
# Load a binary classification dataset (Iris, 2 classes)
iris = datasets.load_iris()
X = iris.data[iris.target < 2]
y = iris.target[iris.target < 2]
```

```python
#Split into train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=0
)
```

```python
#Train SVM with probability estimates enabled
clf = SVC(kernel='linear', probability=True, random_state=0)
clf.fit(X_train, y_train)
```

```python
#Predict probabilities on test set for positive class
y_scores = clf.decision_function(X_test)
```

```python
# Compute PrecisionâRecall and Average Precision (AP)
precision, recall, _ = precision_recall_curve(y_test, y_scores)
ap_score = average_precision_score(y_test, y_scores)
print(f"Average precision (AP): {ap_score:.2f}")
```

```python
 #Plot the Precision-Recall curve
plt.figure(figsize=(8, 6))
plt.step(recall, precision, where='post', color='b', alpha=0.8,
         label=f'AP = {ap_score:.2f}')
plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('PrecisionâRecall Curve for SVM')
plt.legend(loc='lower left')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.grid(True)
plt.show()
```
