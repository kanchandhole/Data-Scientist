### Type I and Type II Errors in Hypothesis Testing

In hypothesis testing, decisions are made based on sample data, and sometimes these decisions can be wrong. These mistakes are called **Type I** and **Type II** errors.

---

### **Type I Error (False Positive)**

**Definition:**
A **Type I error** occurs when the **null hypothesis (Hâ‚€) is true**, but we **reject it** incorrectly.

* Probability of Type I error = **Î± (significance level)**

**Example Scenario:**
A pharmaceutical company tests a new drug.

* Hâ‚€: The drug has **no effect**.
* Hâ‚: The drug **is effective**.

If the test concludes that the drug is effective **when it actually is not**, this is a **Type I error**.

**Real-life meaning:**
Approving an ineffective drug.

---

### **Type II Error (False Negative)**

**Definition:**
A **Type II error** occurs when the **null hypothesis (Hâ‚€) is false**, but we **fail to reject it**.

* Probability of Type II error = **Î²**

**Example Scenario:**
Using the same drug test:

* Hâ‚€: The drug has **no effect**.
* Hâ‚: The drug **is effective**.

If the test concludes that the drug has no effect **when it actually works**, this is a **Type II error**.

**Real-life meaning:**
Rejecting a useful drug.

---

### **Summary Table**

| Error Type  | Reality     | Decision          | Meaning        |
| ----------- | ----------- | ----------------- | -------------- |
| Type I (Î±)  | Hâ‚€ is true  | Reject Hâ‚€         | False Positive |

### **Key Point**

### **Bayesâ€™s Theorem**

**Bayesâ€™s theorem** is a fundamental concept in probability and statistics that helps us **update the probability of an event** based on **new evidence**.

---

### **Mathematical Formula**

[
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
]

Where:

* **P(A)** = Prior probability of event A
* **P(B)** = Probability of event B
* **P(A | B)** = Posterior probability (probability of A given B)
* **P(B | A)** = Likelihood (probability of B given A)

---

### **Simple Explanation**

Bayesâ€™s theorem tells us:

> *â€œHow likely is something after we know some new information?â€*

---

### **Example**

  [
  P(D) = 0.01
  ]

  * Probability test is positive if the person has the disease:
    [
    P(+|D) = 0.99
    ]
  * Probability test is positive if the person does not have the disease:
    [
    P(+|Â¬D) = 0.05
    ]

#### **Question**

---

### **Solution**

First, calculate **P(+)**:

[

**Q5: What is a Confidence Interval? How to calculate it (with example)**

**Confidence Interval (CI):**

* A confidence interval is a range of values used to estimate an unknown population parameter (like mean).
* It tells us how confident we are that the true population value lies within this range.
* Common confidence levels are **90%**, **95%**, and **99%**.

---

### **How to calculate a Confidence Interval (for population mean)**

**Formula (when population standard deviation is known or sample size is large):**
[
\text{CI} = \bar{x} \pm Z \left(\frac{\sigma}{\sqrt{n}}\right)
]

Where:

* (\bar{x}) = sample mean
* (Z) = Z-value (1.96 for 95% CI)
* (\sigma) = standard deviation
* (n) = sample size

---

### **Example:**

**Given:**

* Sample mean ((\bar{x})) = 500
* Standard deviation ((\sigma)) = 50
* Sample size ((n)) = 50
* Confidence level = 95% â†’ (Z = 1.96)

**Step 1: Calculate Standard Error**
[
SE = \frac{50}{\sqrt{50}} = \frac{50}{7.07} \approx 7.07
]

**Step 2: Calculate Margin of Error**
[
ME = 1.96 \times 7.07 \approx 13.86
]

**Step 3: Confidence Interval**
[
CI = 500 \pm 13.86
]

**Final Answer:**

* **Lower limit** = 486.14
* **Upper limit** = 513.86

---

### **Conclusion:**

* We are **95% confident** that the true population mean lies between **486.14 and 513.86**.

[

* Margin of error is **inversely proportional to (\sqrt{n})**.
* As **sample size increases**, (\frac{s}{\sqrt{n}}) decreases, so the **margin of error becomes smaller**.
* Larger samples give **more precise estimates**.

A survey estimates average monthly expenses.

* With **n = 25**:
  [
  \text{MOE} = 1.96 \times \frac{500}{\sqrt{25}} = 196
  ]

* With **n = 100**:
  [
  \text{MOE} = 1.96 \times \frac{500}{\sqrt{100}} = 98
  ]

**Conclusion:**
Increasing the sample size from 25 to 100 **halves the margin of error**, making the estimate more accurate.

**Q12.** A researcher is testing the effectiveness of two different teaching methods on student performance.
Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
significant difference in student performance using a t-test with a significance level of 0.01.

**Ans:**

Letâ€™s solve this **carefully step by step**.

---

## **Given:**

* Sample A: (\bar{x}_1 = 85), (s_1 = 6), (n_1) not given (assume equal, let's call (n_1 = n_2 = n) or we can assume they are large enough for approximation)
* Sample B: (\bar{x}_2 = 82), (s_2 = 5), (n_2) same as above
* Significance level: (\alpha = 0.01)
* Test: **Two-sample t-test (independent samples)**

We are testing whether **the teaching methods have a significant difference**.

---

## **Step 1: State hypotheses**

* Null hypothesis ((H_0)): (\mu_1 = \mu_2) â†’ no difference
* Alternative hypothesis ((H_1)): (\mu_1 \neq \mu_2) â†’ significant difference

This is a **two-tailed t-test**.

---

## **Step 2: Formula for two-sample t-test (unequal variances)**

[
t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
]

We also need **degrees of freedom** (Welch-Satterthwaite formula):

[
df = \frac{\left(\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}\right)^2}{\frac{(s_1^2/n_1)^2}{n_1-1} + \frac{(s_2^2/n_2)^2}{n_2-1}}
]

---

Since **sample sizes are not given**, letâ€™s assume **n1 = n2 = 30** (common assumption for t-tests if not specified).

---

## **Step 3: Compute standard error (SE)**

[
SE = \sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}
]

[
s_1^2 = 6^2 = 36, \quad s_2^2 = 5^2 = 25
]

[
SE = \sqrt{\frac{36}{30} + \frac{25}{30}} = \sqrt{1.2 + 0.8333} = \sqrt{2.0333} \approx 1.426
]

---

## **Step 4: Compute t-statistic**

[
t = \frac{\bar{x}_1 - \bar{x}_2}{SE} = \frac{85 - 82}{1.426} = \frac{3}{1.426} \approx 2.105
]

---

## **Step 5: Determine degrees of freedom (df)**

[
df = \frac{(s_1^2/n_1 + s_2^2/n_2)^2}{(s_1^2/n_1)^2/(n_1-1) + (s_2^2/n_2)^2/(n_2-1)}
]

Step by step:

* Numerator: ( (1.2 + 0.8333)^2 = (2.0333)^2 \approx 4.134 )

* Denominator:
  [
  \frac{(1.2)^2}{29} + \frac{(0.8333)^2}{29} = \frac{1.44}{29} + \frac{0.6944}{29} = 0.04966 + 0.02394 \approx 0.0736
  ]

[
df \approx \frac{4.134}{0.0736} \approx 56.2 \approx 56
]

---

## **Step 6: Find critical t-value**

* Two-tailed test, Î± = 0.01 â†’ Î±/2 = 0.005 in each tail
* df â‰ˆ 56 â†’ (t_{critical} \approx 2.66)

---

## **Step 7: Compare t-statistic with critical value**

[
t_{calculated} = 2.105 < t_{critical} = 2.66
]

* Since (t < t_{critical}), we **fail to reject the null hypothesis**.

---

## **Step 8: Conclusion**

At the **1% significance level**, there is **not enough evidence** to say that the two teaching methods have a significant difference in student performance.

## **Given:**

---

## **Step 1: State hypotheses**

* Alternative hypothesis ((H_1)): (\mu \neq 0.30) â†’ caffeine **affects reaction time**

## **Step 2: Compute the t-statistic**

[

]

[

[

]

---

### Step 2b: Compute t

[
t = \frac{-0.05}{0.00913} \approx -5.47
]

* Two-tailed test, Î± = 0.10 â†’ Î±/2 = 0.05 per tail
