**Q1.** What is an API? Give an example, where an API is used in real life.

**Ans:**API is a software interface that allows two applications to communicate with each other; for example, a weather app uses a weather API to get live temperature data.

**Q2.** Give advantages and disadvantages of using API.

**ans:** Advantages: APIs enable automation, integration, and faster development; Disadvantages: they depend on network reliability and may have security or versioning issues.

**Q3.** What is a Web API? Differentiate between API and Web API.

**ans:** A Web API is an API accessed over the web using HTTP; API is a general interface, while Web API specifically works through the internet.

**Q4. **Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

**ans:** REST is a lightweight architectural style using HTTP, whereas SOAP is a strict protocol using XML; SOAP is slower, more complex, and less flexible.

**Q5.** Differentiate between REST and SOAP.

**ans:** REST is simple, fast, and uses multiple data formats, while SOAP is heavyweight, XML-based, and highly structured.
