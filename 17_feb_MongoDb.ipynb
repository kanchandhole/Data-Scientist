**Q1.** What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?

**Ans:**

MongoDB is a NoSQL, document-oriented database that stores data in JSON-like documents (BSON).

Non-relational databases store data without fixed tables and schemas, allowing flexible and dynamic data structures.

MongoDB is preferred over SQL when:

Data structure is unstructured or semi-structured

Schema changes frequently

High scalability and performance are required

Handling large volumes of data (Big Data)

Real-time applications (chat apps, IoT, analytics)

**Q2.** State and Explain the features of MongoDB.

**Ans:**

Schema-less: No fixed table structure

Document-based: Stores data in BSON format

High scalability: Supports horizontal scaling

High performance: Faster read/write operations

Indexing: Improves query performance

Replication: High availability via replica sets

Aggregation framework: Advanced data processing

**Q3.** Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.

```python
!pip install pymongo
```

```python
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Create database
db = client["school"]

# Create collection
collection = db["students"]

print("Database and Collection created")
```

**Q4.** Using the database and the collection created in question number 3, write a code to insert one record,
and insert many records. Use the find() and find_one() methods to print the inserted record.

**Ans:**

#Insert One Record
collection.insert_one({
    "name": "Kanchan",
    "age": 22,
    "course": "Python"
})


#Insert Many Records

collection.insert_many([
    {"name": "Amit", "age": 21, "course": "Data Science"},
    {"name": "Neha", "age": 23, "course": "AI"}
])

#Find One Record

print(collection.find_one())


#Find All Records

for record in collection.find():
    print(record)


**Q5.** Explain how you can use the find() method to query the MongoDB database. Write a simple code to
demonstrate this.

**Ans:**The find() method retrieves documents based on conditions.

# Find students enrolled in Python
for record in collection.find({"course": "Python"}):
    print(record)

**Q6.** Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

**Ans:**

The sort() method sorts documents in ascending (1) or descending (-1) order.

# Sort by age in ascending order
for record in collection.find().sort("age", 1):
    print(record)

**Q7.** Explain why delete_one(), delete_many(), and drop() is used.

**Ans:**

| Method          | Purpose                                       |
| --------------- | --------------------------------------------- |
| `delete_one()`  | Deletes **one matching document**             |
| `delete_many()` | Deletes **all matching documents**            |
| `drop()`        | Deletes the **entire collection or database** |



collection.delete_one({"name": "Amit"})

collection.delete_many({"course": "Python"})

collection.drop()
