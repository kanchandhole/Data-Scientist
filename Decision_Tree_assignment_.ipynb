**Q1.**What is a Decision Tree, and how does it work?

**Ans:** A decision tree is a flowchart-like structure used in machine learning and decision analysis to model possible outcomes and decisions. It's a supervised learning algorithm that breaks down complex problems into a series of simpler decisions based on attributes or features. Each node in the tree represents a test or a decision, branches represent the possible outcomes of the test, and leaf nodes represent the final outcome or prediction.

How it works:
1. Training Data:
The algorithm learns from a dataset containing past outcomes and related variables (features).
2. Root Node:
The process starts with the root node, representing the entire dataset.
3. Splitting:
The algorithm identifies the most important feature to split the data based on, creating subtrees.
4. Decision Rules:
Each internal node represents a decision based on a specific feature, and branches represent the different possible values of that feature.
5. Leaf Nodes:
When the data is split into homogeneous groups (e.g., all outcomes of one type), the process reaches the leaf nodes, which represent the final decisions or predictions.
6. Predicting:
To use the tree for prediction, you start at the root and follow the branches based on the input data, eventually reaching a leaf node that provides the prediction.

**Q2.**What are impurity measures in Decision Trees?

**Ans:** Impurity measures in decision trees quantify the degree of homogeneity or "purity" of a set of data instances within a node. They help determine which features to use for splitting the tree, aiming to create more homogeneous subsets at each level.

Common Impurity Measures:

Gini Impurity:
This measure calculates the probability that a randomly chosen element from the node would be incorrectly classified. A lower Gini impurity indicates a more pure node.

Entropy:
Entropy measures the disorder or randomness in a dataset. A lower entropy value indicates a more pure node.

Misclassification Error:
This measure calculates the proportion of incorrect predictions within a node. A lower misclassification error suggests a more pure node.

How Impurity Measures are Used:

Splitting Criteria:
Decision tree algorithms use impurity measures to evaluate how well a particular feature splits a node into child nodes. The algorithm selects the split that results in the greatest reduction in impurity.

Node Evaluation:
Impurity measures are used to assess the purity of a node. A pure node contains instances belonging to a single class, while an impure node contains a mix of

classes.
Algorithm Selection:
Some decision tree algorithms use specific impurity measures, while others allow for the choice of measure.

**Q3**.What is the mathematical formula for Gini Impurity.

**Ans:** The Gini Impurity formula is Gini(D) = 1 - Σ pi^2, where 'i' represents different classes and 'pi' is the probability of a data point belonging to class 'i'.

**Q4.**What is the mathematical formula for Entropy?

**Q5.**What is Information Gain, and how is it used in Decision Trees?

**Ans:** Information Gain (IG) is a key metric in decision tree algorithms, representing the reduction in uncertainty or entropy about a target variable when a particular feature is known. It helps determine the best attribute to split a node in a decision tree, ensuring that the tree is built efficiently and accurately.

Entropy:
Before splitting, the data in a node might be mixed (high entropy) if it contains instances of multiple classes. Entropy quantifies this impurity or uncertainty.

Information Gain Calculation:
Information Gain is calculated by subtracting the weighted average entropy of the child nodes from the parent node's entropy.

Splitting Decision:
The algorithm selects the feature that results in the highest Information Gain, meaning it most effectively reduces the uncertainty in the data after the split.

Example:
Imagine a decision tree trying to predict whether a restaurant customer will return. The "patrons" attribute (e.g., many, some, none) might have a higher Information Gain than "type of food" because splitting on "patrons" leads to a more predictable classification of return customers.

**Q6.**What is the difference between Gini Impurity and Entropy.

**Ans:** Both entropy and Gini impurity are used as splitting criteria in decision trees, but they have slight differences. While entropy tends to create more balanced trees, Gini impurity tends to isolate the most frequent class in its own branch of the tree.

**Q8.**What is Pre-Pruning in Decision Trees?

the growth of the decision tree can be stopped before it gets too complex, this is called pre-pruning. It is important to prevent the overfitting of the training data, which results in a poor performance when exposed to new data.

**Q14.**How does a Decision Tree handle categorical features.

**Ans:** Decision Trees handle categorical features:
1. Theoretical Ability:
Decision Trees are designed to work with both numerical and categorical data.
The algorithm can determine the best splits at each node based on the information gain from splitting on categorical features.
2. Preprocessing for Implementation:
One-Hot Encoding:
Transforms each category into a binary vector. Each category gets its own column, with a 1 in that column if the sample belongs to that category and 0 otherwise. This is useful when there's no inherent order among categories.
Label Encoding:
Assigns a unique numerical label (integer) to each category. This is suitable when there's an ordinal relationship between the categories.
Other Encoding Methods:
Other techniques, like Binary Encoding or Mean Encoding, can also be used to convert categorical features into numerical form.
3. Handling Categorical Features Directly:
Some implementations of Decision Trees (like XGBoost and LightGBM) can handle categorical features without explicit preprocessing.
These implementations might use gradient statistics to convert categorical features to numerical values for splitting.
4. Considerations:
Large Number of Categories:
For a large number of categories, one-hot encoding can lead to a high-dimensional feature space. Dimensionality reduction techniques may be needed to avoid sparse matrices.
Missing Values:
Missing values in categorical features can be handled by imputing them (e.g., using the most frequent category) or by treating them as a separate category.
Unseen Categories:
If the model encounters categories in the test set that weren't present in the training set, they need to be handled appropriately (e.g., by ignoring them, replacing them with a special value, or raising an error)

```python
clf_scaled = DecisionTreeClassifier(random_state=42)
clf_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = clf_scaled.predict(X_test_scaled)
acc_scaled = accuracy_score(y_test, y_pred_scaled)
```

Q27. Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and
visualize its effect on accuracy.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load dataset
iris = load_iris()
X, y = iris.data, iris.target
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train an initial tree to get effective alphas
clf = DecisionTreeClassifier(random_state=0)
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas
```

```python
# Train a tree for each ccp_alpha
clfs = []
for alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)
```

```python
# Remove the last tree (usually underfits heavily)
clfs = clfs[:-1]
ccp_alphas = ccp_alphas[:-1]
```

```python
# Evaluate accuracy
train_acc = [accuracy_score(y_train, clf.predict(X_train)) for clf in clfs]
test_acc = [accuracy_score(y_test, clf.predict(X_test)) for clf in clfs]
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, train_acc, marker='o', label='Train Accuracy', drawstyle="steps-post")
plt.plot(ccp_alphas, test_acc, marker='o', label='Test Accuracy', drawstyle="steps-post")
plt.xlabel("ccp_alpha (Pruning Strength)")
plt.ylabel("Accuracy")
plt.title("Effect of Cost Complexity Pruning on Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**30.**Write a Python program to train a Decision Tree Classifier and use GridSearchCV to find the optimal values
for max_depth and min_samples_split.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Define the parameter grid
param_grid = {
    'max_depth': [2, 3, 4, 5, 6, None],
    'min_samples_split': [2, 4, 6, 8, 10]
}
```

```python

# Set up the Decision Tree and GridSearchCV
clf = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='accuracy')
```

```python
# Fit the model
grid_search.fit(X_train, y_train)
```

```python
# Best parameters and estimator
print("🔧 Best Parameters:", grid_search.best_params_)
print("✅ Best Cross-Validation Accuracy: {:.2f}%".format(grid_search.best_score_ * 100))
```

```python
# Evaluate on test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

print("\n📊 Classification Report on Test Set:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```
