**Q1.** Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

**Ans:** A scatter plot is used to show the relationship between two numerical variables.

```python
import seaborn as sns
import plotly.express as px

# Load dataset
titanic = sns.load_dataset("titanic")

# Scatter plot
fig = px.scatter(titanic, x="age", y="fare",
                 title="Scatter Plot of Age vs Fare (Titanic Dataset)")
fig.show()
```

**Q2.** Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

**Ans:** A box plot shows the distribution of data and highlights median, quartiles, and outliers.

```python
import plotly.express as px

tips = px.data.tips()

fig = px.box(tips, x="day", y="total_bill",
             title="Box Plot of Total Bill by Day")
fig.show()
```

**Q3.** Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

**Ans:**

* A histogram shows frequency distribution.

* pattern_shape differentiates smoker status

color differentiates days

```python
import plotly.express as px

tips = px.data.tips()

fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker",
    title="Histogram of Total Bill by Sex"
)
fig.show()
```

**Q4.** Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

**Ans:** A scatter matrix shows pairwise relationships between multiple numerical variables.

```python
import plotly.express as px

iris = px.data.iris()

fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width",
                "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset"
)
fig.show()
```

**Q5.** What is Distplot? Using Plotly express, plot a distplot.

**Ans:**

A Distplot (Distribution Plot) is used to visualize:

Histogram

Probability density curve (KDE)

Note:
distplot is deprecated in Seaborn. In Plotly, similar behavior is achieved using histogram with marginal="rug" or marginal="box".

```python
import plotly.express as px
import numpy as np

data = np.random.randn(1000)

fig = px.histogram(
    data,
    nbins=30,
    marginal="rug",
    title="Distplot using Plotly Express"
)
fig.show()
```
