**Q1.** What is multiprocessing in python? Why is it useful?

**Ans:** Multiprocessing is a Python technique that allows running multiple processes simultaneously, each with its own memory space.
It is useful for CPU-bound tasks and helps bypass Pythonâ€™s GIL (Global Interpreter Lock) to achieve true parallelism.

**Q2.** What are the differences between multiprocessing and multithreading?

**ans:**

| Aspect    | Multiprocessing              | Multithreading   |
| --------- | ---------------------------- | ---------------- |
| Execution | Multiple processes           | Multiple threads |
| Memory    | Separate memory              | Shared memory    |
| GIL       | Not affected                 | Affected by GIL  |
| Best for  | CPU-bound tasks              | I/O-bound tasks  |
| Speed     | Faster for heavy computation | Faster for I/O   |


**Q3.** Write a python code to create a process using the multiprocessing module.

**Ans:**

```python
import multiprocessing

def task():
    print("Process is running")

if __name__ == "__main__":
    p = multiprocessing.Process(target=task)
    p.start()
    p.join()
```

**Q4**. What is a multiprocessing pool in python? Why is it used?

**ans:**

A multiprocessing pool manages a group of worker processes and distributes tasks among them automatically.
It is used to reuse processes efficiently and reduce process creation overhead.

**Q5.** How can we create a pool of worker processes in python using the multiprocessing module?

**ans:**

```python
from multiprocessing import Pool

def square(n):
    return n * n

if __name__ == "__main__":
    with Pool(4) as p:
        result = p.map(square, [1, 2, 3, 4])
    print(result)
```

**Q6**. Write a python program to create 4 processes, each process should print a different number using the
multiprocessing module in python.

**Ans:**

```python
import multiprocessing

def print_number(num):
    print("Process number:", num)

if __name__ == "__main__":
    processes = []

    for i in range(4):
        p = multiprocessing.Process(target=print_number, args=(i,))
        processes.append(p)
        p.start()

    for p in processes:
        p.join()
```
