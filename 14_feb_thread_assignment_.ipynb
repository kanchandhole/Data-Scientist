**Q4.** Write a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes.

```python
import time
import threading
def print_squares(n):
  for i in n:
    print(f" squares of value {i}", (i*i))
def print_cubes(n):
  for i in n:
    print(f"cubes of value {i}" , (i*i*i))
```

```python
num = [1,2,3,4,5]
```

```python
# Creating threads
t1 = threading.Thread(target=print_squares, args=(num,))
t2 = threading.Thread(target=print_cubes, args=(num,))
```

```python
t1.start()
t2.start()
```

**Q5.** State advantages and disadvantages of multithreading.

**ans:** Advantages of Multithreading

Improved Performance
Multiple tasks run concurrently, reducing overall execution time.

Better CPU Utilization
CPU remains active while threads wait for I/O operations.

Increased Responsiveness
Applications remain responsive (e.g., UI, servers).

Efficient Resource Sharing
Threads share memory and resources, reducing overhead.

Faster Context Switching
Switching between threads is faster than processes.

 Disadvantages of Multithreading

Complex Programming
Writing and debugging multithreaded code is difficult.

Race Conditions
Multiple threads accessing shared data can cause errors.

Deadlocks
Threads may wait indefinitely for resources.

Synchronization Overhead
Locks and synchronization slow performance.

GIL Limitation in Python
Python‚Äôs Global Interpreter Lock limits CPU-bound multithreading.

**Q6.** Explain deadlocks and race conditions.

Deadlock

‚úÖ Definition

A deadlock occurs when two or more threads are waiting indefinitely for resources held by each other, so none of them can proceed.

üîπ Why Deadlock Happens

Multiple threads hold resources

Each thread waits for a resource held by another

No thread releases its resource

üîπ Simple Example (Real-life)

Two people holding:

Person A ‚Üí key

Person B ‚Üí lock
Both wait for the other ‚Üí no progress ‚ùå

üîπ Python Example
lock1 = threading.Lock()
lock2 = threading.Lock()

# Thread 1 acquires lock1 then waits for lock2
# Thread 2 acquires lock2 then waits for lock1


‚û° Program freezes.

üîπ How to Prevent Deadlock

‚úî Acquire locks in the same order

‚úî Use timeout in locks

‚úî Avoid unnecessary locking

‚ö†Ô∏è Race Condition

‚úÖ Definition

A race condition occurs when multiple threads access and modify shared data simultaneously, causing unpredictable and incorrect results.

üîπ Why Race Condition Happens

Shared variable

No synchronization

Threads execute at the same time

üîπ Python Example
count = 0

def increment():
    global count
    count += 1


Multiple threads ‚Üí incorrect value ‚ùå

üîπ How to Prevent Race Condition

‚úî Use locks (threading.Lock)

‚úî Use thread-safe data structures

‚úî Minimize shared data
