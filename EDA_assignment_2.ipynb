Q9.How do car prices vary based on the customer's annual income bracket.

```python
bins = [100, 1000, 2000, 4000, 8000, 12000, float('inf')]
labels = ['100–1k$', '1k–2k$', '2k–4k$', '4k–8k$', '8k–12k$', '12k$+']
df_copy['Income_Bracket'] = pd.cut(df_copy['Annual Income'], bins=bins, labels=labels)
```

```python
avg_price_by_bracket = df_copy.groupby('Income_Bracket')['Price ($)'].mean()
```

```python
avg_price_by_bracket
```

```python
avg_price_by_bracket.plot(kind='bar', color='teal', figsize=(7,4))
plt.title('Average Car Price by Annual Income Bracket')
plt.xlabel('Annual Income Bracket')
plt.ylabel('Average Selling Price (USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Q12.**Is there any seasonal trend in car sales based on the date of sale.

```python
df_copy['Month'] = df_copy['Date'].dt.month
df_copy['Year'] = df_copy['Date'].dt.year
df_copy['Month_Name'] = df_copy['Date'].dt.strftime('%B')
```

```python
print(df_copy.head())
print(df_copy['Month'].value_counts().sort_index())
```

```python
monthly_sales = df_copy.groupby('Month')['Date'].count()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette='viridis')
plt.title('Monthly Car Sales Trend')
plt.xlabel('Month')
plt.ylabel('Number of Sales')
plt.xticks(ticks=range(0,12), labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.tight_layout()
plt.show()
```
