**Q1.**What is the average rating for each product category.

```python
avg_rating_category = df_cleaned.groupby('category')['rating'].mean()
```

```python
plt.figure(figsize=(12,6))
avg_rating_category.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Average Rating for Each Product Category')
plt.xticks(rotation=90)
plt.show()
```

**Q2.**What are the top rating_count products by category.

```python
top_rating_count_product = df_cleaned.groupby('category')['rating_count'].value_counts().sort_values(ascending=False)
```

```python
top_rating_count_product.head(5)
```

**Q3.**What is the distribution of discounted prices vs. actual prices.

```python
plt.figure(figsize=(10,6))

sns.kdeplot(df_cleaned['actual_price'], label='Actual Price', fill=True, color='blue')
sns.kdeplot(df_cleaned['discounted_price'], label='Discounted Price', fill=True, color='green')

plt.title('Distribution: Discounted vs Actual Prices')
plt.xlabel('Price')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```
