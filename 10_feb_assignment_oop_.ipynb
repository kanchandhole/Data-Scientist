**Q1.** Which function is used to open a file? What are the different modes of opening a file? Explain each mode
of file opening.

**ANS:** In Python, the open() function is used to open a file.

file_object = open(filename, mode)

Mode	Meaning	Description

'r'	Read	Opens the file for reading (default). File must already exist.

'w'	Write	Opens the file for writing. If file exists, it is overwritten. If it doesn’t exist, it is created.

'a'	Append	Opens the file for writing, but adds new data at the end instead of overwriting.

'r+'	Read and Write	Opens file for both reading and writing. File must already exist.

'w+'	Write and Read	Opens file for writing and reading. Overwrites existing file or creates new one.

'a+'	Append and Read	Opens file for reading and appending. File pointer is at the end; existing data is not overwritten.

'x'	Exclusive Creation	Creates a new file; raises an error if file already exists.

Mode	Description

'rb'	Read binary file

'wb'	Write binary file

'ab'	Append binary file

**Q2.** Why close() function is used? Why is it important to close a file?

**Ans:** The close() function in Python is used to close a file that was opened using the open() function.

Purpose of close()

When you open a file with open(), the file is connected to your program through a file object.
Python keeps this connection active in memory until you explicitly close it.

So, the close() method:

Releases system resources associated with the file

Flushes (saves) any data still in the buffer

Prevents accidental changes after work is done

```python
file = open("example.txt", "w")
# Perform operations
file.close()
```

**Q4.** Explain the following with python code: read(), readline() and readlines().

```python
f = open("data_scientist.txt", "r") #The read() method is used to read the entire content of a file (or a specific number of characters) as a single string.
print(f.read())
```

```python
f.close()
```

```python
f = open("data_scientist.txt", "r") #The readline() method reads one line at a time from a file.
print(f.readline())
```

```python
f = open("data_scientist.txt", "r") #The readlines() method is used to read all lines from a text file and return them as a list of strings,where each string represents one line from the file
print(f.readlines())
```

**Q5.** Explain why with statement is used with open(). What is the advantage of using with statement and
open() together?

**Ans**: The with statement in Python is used to open and automatically close a file after the block of code inside it is executed — even if an error occurs.

```python
with open("file.txt", "w") as f:
  f.write("new file")
```

```python
#Q6. Explain the write() and writelines() functions. Give a suitable example.

f = open("file.txt", "w")
f.write("new file") # it write single line
f.close()
```
