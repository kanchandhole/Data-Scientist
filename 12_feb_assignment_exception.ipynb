**Q1.** What is an Exception in python? Write the difference between Exceptions and Syntax errors.

**Ans:** An exception is an error that occurs during program execution; syntax errors occur before execution due to incorrect code structure.

**Q2.**What happens when an exception is not handled? Explain with an example.

**Ans:** If an exception is not handled, the program stops immediately and prints a traceback error (e.g., 1/0 crashes the program).

**Q3.**Which Python statements are used to catch and handle exceptions? Explain with an example.

**Ans:** Exceptions in Python are caught and handled using try and except statements.

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("You cannot divide by zero!")
```

**Q4.** Explain with an example:

**ans:**

1️⃣ try

Used to wrap code where an exception may occur.

2️⃣ else

Runs only if no exception occurs in the try block.

3️⃣ finally

Runs always, whether an exception happens or not.

4️⃣ raise

Used to manually throw an exception.

try: runs the code and hits raise ValueError(...)

raise: throws an exception manually

except: catches the error and prints it

else: skipped because an exception occurred

finally: runs regardless of the exception

```python
try:
    num = 10
    if num == 10:
        raise ValueError("Number cannot be 10")   # raise
except ValueError as e:
    print("Handled Error:", e)
else:
    print("Executed only if no exception occurs")
finally:
    print("This will run no matter what")
```

**Q5.**What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example.

**ans:** Custom exceptions are user-defined error classes created to handle specific application-related errors that Python’s built-in exceptions cannot describe clearly.

why we need custom exception:

We need custom exceptions to make error messages more meaningful, improve debugging, and enforce rules specific to the program’s logic.



**Q6.**Create a custom exception class. Use this class to handle an exception.

```python
class AgeError(Exception):
    pass

def check_age(age):
    if age < 18:
        raise AgeError("Age must be 18 or above!")
    else:
        print("Valid age")

try:
    check_age(15)
except AgeError as e:
    print("Custom Exception Caught:", e)
```

**Q6.** Create a custom exception class. Use this class to handle an exception.

```python
#Custom Exception Class
class InvalidMarksError(Exception):
    pass
```

```python
#Function Using the Exception
def check_marks(marks):
    if marks < 0 or marks > 100:
        raise InvalidMarksError("Marks must be between 0 and 100!")
    else:
        print("Marks are valid.")
```

```python
#Handling the Exception
try:
    check_marks(150)
except InvalidMarksError as e:
    print("Custom Exception Caught:", e)
```
