**Q10.**What is the relationship between selling_price and km_driven for first-owner bikes

```python
df.head()
```

```python
first_owner_bikes = df[df['owner'].str.lower().str.strip() == '1st owner']
```

```python
# Drop rows with missing price or km values
first_owner_bikes = first_owner_bikes.dropna(subset=['selling_price', 'km_driven'])
```

```python
# Plot relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(data=first_owner_bikes, x='km_driven', y='selling_price', alpha=0.5)
sns.regplot(data=first_owner_bikes, x='km_driven', y='selling_price', scatter=False, color='red')
plt.title('Relationship Between Selling Price and KM Driven (First Owner Bikes)')
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price')
plt.tight_layout()
plt.show()
```

**Q11.**Identify and remove outliers in the km_driven column using the IQR method

```python
df.head()
```

```python
minimum,Q1,median,Q3,maximum=np.quantile(df['km_driven'],[0,0.25,0.50,0.75,1.0])
```

```python
minimum,Q1,median,Q3,maximum
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
lower_fence=Q1-1.5*(IQR)
higher_fence=Q3+1.5*(IQR)
```

```python
lower_fence
```

```python
higher_fence
```

```python
df_no_outliers = df[(df['km_driven'] >= lower_fence) & (df['km_driven'] <= higher_fence)]
```

```python
df_no_outliers
```

```python
removed = len(df) - len(df_no_outliers)
```

```python
removed
```

```python
plt.figure(figsize = (10,9))
sns.boxplot(x=df['km_driven'])
plt.title("Before Remove Outlier")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,9))
sns.boxplot(df_no_outliers)
plt.title("After Remove Outlier")
plt.tight_layout()
plt.show()
```
