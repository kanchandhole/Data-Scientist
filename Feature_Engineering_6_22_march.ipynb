Q2. How did you handle missing data in the wine quality data set during the feature engineering process?
Discuss the advantages and disadvantages of different imputation techniques.

**Ans:**



## 🔍 1. How Missing Data is Handled

### Step 1: Check for Missing Values

* Using:

  * `df.isnull().sum()`
* If **no missing values** → no action needed
* If present → apply imputation

---

## 🛠️ 2. Common Imputation Techniques

### ✅ 1. Mean Imputation

* Replace missing values with **mean of the column**

**Example:**
Alcohol missing → replace with average alcohol value

**Advantages:**

* Simple and fast
* Works well for **normally distributed data**

**Disadvantages:**

* Affected by outliers
* Reduces data variability

---

### ✅ 2. Median Imputation

* Replace with **median value**

**Advantages:**

* Robust to outliers
* Best for **skewed data**

**Disadvantages:**

* May ignore actual relationships between features

---

### ✅ 3. Mode Imputation

* Used for **categorical features**

**Advantages:**

* Simple
* Maintains most frequent category

**Disadvantages:**

* Can introduce bias

---

### ✅ 4. Forward Fill / Backward Fill

* Fill using previous/next values

**Advantages:**

* Useful for **time-series data**

**Disadvantages:**

* Not suitable for independent samples like wine data

---

### ✅ 5. KNN Imputation

* Uses nearest neighbors to estimate missing values

**Advantages:**

* More accurate
* Considers feature relationships

**Disadvantages:**

* Computationally expensive
* Sensitive to scaling

---

### ✅ 6. Regression Imputation

* Predict missing values using other features

**Advantages:**

* Captures relationships
* More precise

**Disadvantages:**

* Complex
* Risk of overfitting

---

## 📊 3. What You Should Write for Wine Dataset

👉 Since Wine Quality dataset usually has **no missing values**:

**Best Answer (Exam Ready):**

> “The dataset was first checked for missing values using `.isnull().sum()`. Since no missing values were found, no imputation was required. However, if missing values were present, median imputation would be preferred due to robustness against outliers in chemical properties.”

---

## ✅ 4. Final Conclusion

* Always **check first**, then decide technique
* For wine dataset:

  * Median → best practical choice
  * KNN/Regression → advanced approaches

# 🔧 3. Transformations to Improve Normality

---

## ✅ 1. Log Transformation

Best for **right-skewed data**

```python
import numpy as np
df['residual sugar'] = np.log1p(df['residual sugar'])
```

👉 Apply to:

* residual sugar
* chlorides
* sulfur dioxide

---

## ✅ 2. Square Root Transformation

```python
df['chlorides'] = np.sqrt(df['chlorides'])
```

👉 Less aggressive than log

---

## ✅ 3. Box-Cox Transformation

```python
from scipy.stats import boxcox
df['sulphates'], _ = boxcox(df['sulphates'] + 1)
```

👉 Works only for positive data
👉 Very effective for normalization

---

## ✅ 4. Yeo-Johnson Transformation

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')
df[['feature']] = pt.fit_transform(df[['feature']])
```

👉 Works for both positive & negative values

---

# 📌 4. Final Answer (Most Important Part)

### Features with Non-Normal Distribution:

* Residual sugar
* Chlorides
* Free sulfur dioxide
* Total sulfur dioxide
* Sulphates
* Volatile acidity

---

### Recommended Transformations:

* **Log / sqrt** → for skewed features
* **Box-Cox / Yeo-Johnson** → for strong normalization

---

# ✅ 5. Conclusion

👉 Many chemical features are **skewed**, which can:

* affect model performance
* violate assumptions (like linear regression)

👉 Applying transformations:

* improves **normality**
* enhances **model accuracy**

---



> Several features such as residual sugar, chlorides, and sulfur dioxide are right-skewed. Log, square root, and Box-Cox transformations can be applied to normalize these features and improve model performance.

**Q6.** Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?

**Ans:**

After applying PCA on the dataset, approximately 4–6 principal components are expected to explain 90% of the variance, reducing dimensionality while preserving most of the information.

````python
#Step 1: Load Data

import pandas as pd

df = pd.read_csv("Umazon - Case Base Data.xlsx - base data.csv")

````

````python
df.head(5)
````

````python
df['Departments'].unique()
````

````python
df['Buyer_Category'].unique()
````

````python
df['Merchant_Category'].unique()
````

````python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Departments'] = le.fit_transform(df['Departments'])
````

````python
df['Buyer_Category'] =le.fit_transform(df['Buyer_Category'])
````

````python
df['Merchant_Category'] = le.fit_transform(df['Merchant_Category'])
````

````python
df.head()
````

````python
X = df.drop(columns=[])  # (no target given, so use all features)
````

````python
#Step 2: Standardization (IMPORTANT)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python
#Step 3: Apply PCA

from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(X_scaled)
````

````python
#Step 4: Explained Variance

import numpy as np

explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

print(cumulative_variance)
````

````python
#Step 5: Find Components for 90%

n_components_90 = np.argmax(cumulative_variance >= 0.90) + 1
print("Components needed for 90% variance:", n_components_90)
````

````python
import matplotlib.pyplot as plt

plt.plot(cumulative_variance, marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Scree Plot")
plt.show()
````
